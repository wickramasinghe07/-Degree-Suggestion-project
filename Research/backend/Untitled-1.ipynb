```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Drop non-numeric columns before scaling
numeric_df = df.drop(columns=['Degree_Program'])

# Standardize the data
# scaler = StandardScaler()
# scaled_data = scaler.fit_transform(numeric_df)

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
principal_components = pca.fit_transform(numeric_df)

# Perform KMeans clustering
kmeans = KMeans(n_clusters=6, random_state=42)
kmeans.fit(principal_components)
cluster_labels = kmeans.labels_

# Plot the clusters
plt.figure(figsize=(8, 6))
plt.scatter(principal_components[:, 0], principal_components[:, 1], c=cluster_labels, cmap='viridis', alpha=0.5)
plt.title('KMeans Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Initialize a set to store unique predicted labels
unique_labels = set()

# Counter for the number of unique predicted labels
unique_count = 0

# Loop until we collect 6 different predicted labels
while unique_count < 6:
    # Generate a random test array
    testarry = np.random.randint(0, 4, size=(4,))
    
    # Reshape the array for prediction
    input_array = np.array(testarry).reshape(1, -1)
    
    # Perform predictions using all loaded models
    predictions = [classifier.predict(input_array)[0] for classifier in classifiers]
    ensemble_prediction = max(set(predictions), key=predictions.count)

    predicted_label = disorder_labels.get(ensemble_prediction, "Unknown Disorder")
    
    # Check if the predicted label is unique
    if predicted_label not in unique_labels:
        unique_labels.add(predicted_label)
        unique_count += 1
        
        # Print the predicted label and the input array if it's unique
        print("Predicted Label:", predicted_label)
        print("Input Array:", input_array)
```
